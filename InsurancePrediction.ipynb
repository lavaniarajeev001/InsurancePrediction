```python
X=df.iloc[: , 1:9]
```

```python
y=df.TravelInsurance
```

```python
X["Employment Type"].value_counts(),X["GraduateOrNot"].value_counts(),X["FrequentFlyer"].value_counts(),X["EverTravelledAbroad"].value_counts()
```

```python
ET={"Private Sector/Self Employed":1,"Government Sector":2}
GT={"Yes":1,"No":2}
FF={"Yes":1,"No":2}
TA={"Yes":1,"No":2}
```

```python
X["Employment Type"]=X["Employment Type"].map(ET)
X["GraduateOrNot"]=X["GraduateOrNot"].map(GT)
X["FrequentFlyer"]=X["FrequentFlyer"].map(FF)
X["EverTravelledAbroad"]=X["EverTravelledAbroad"].map(TA)
X.head()
```

```python
from imblearn.combine import SMOTETomek
```

```python
smk=SMOTETomek(random_state=42)
X_res,y_res=smk.fit_resample(X,y)
```

```python
X_res.shape,y_res.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_res,y_res,test_size=0.30,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier=RandomForestClassifier()
classifier.fit(X_train,y_train)
```
